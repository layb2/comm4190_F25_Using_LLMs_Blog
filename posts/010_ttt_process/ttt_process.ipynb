In the TikTok Task course project, we were tasked with exploring the role that TikTok plays in forming public opinion. Social media platforms like TikTok have become popular places for discourse to occur, and the modern debate of remote vs in-person work is no exception. 

This blog post will feature a high-level breakdown of the methods used in the project, and will be part of a larger series that dives deeper into aspects of the process and the results.

1. To start, I defined a research question: do users who engage with the comment section of this TikTok video prefer remote or in-person/in-office work?
2. Next, I wrote a prompt for an LLM to code comments based on that research question. In beginning the human coding according to that research question/concepts, I discovered that there were almost no comments that supported in-office only, and I altered the research question and concepts accordingly.
3. Next, I began redoing the human coding according to the new research question, which instead studied preferences for hybrid vs non-hybrid work (whether remote or in-office), and the revised concepts that I had developed.
4. Then I ran the code for the LLM to code comments. The agreement between my coding and that of the LLM was unsatisfactory, so I began a time-intensive process of revising the concept wordings to have clearer distinctions between categories and better capture specific types of minorities of comments that the LLM was consistently getting wrong.
5. When the agreement was finally satisfactory, I ran the code to create the graphs, created one more that better contextualized the previous figures, and analyzed the results. Read more about this in the future blog post.
