While working on this LLM-assisted study of sentiments surrounding hybrid vs non-hybrid work in the comment section of a TikTok video, improving the inter-rater reliability score was the most time-intensive part. 

Minor changes in wording either harmed one concepts score whilst improving the other, caused minimal improvements to both, or had significant negative effects on both scores.

A time-intensive process of trial and error, combined with focusing on being as clear and concise as possible while still addressing the variety of possible comment types, was most effective in improving the scores.

For example, providing the LLM with examples of keywords that signaled preferences was helpful in improving the score. The scores started to near the 0.7 threshold, but my refinements stopped making much improvement, so I examined the LLM-coded dataset to see if there was a certain type of comment that it was miscoding.

After identifying that it struggled to code comments where agreement was expressed but followed by a statement of preference, I updated the concepts (e.g. Code 0 ifthe agreement is not clearly about the videos advocacy for hybrid.) in order to better handle that type of comment and create a clearer distinction between concepts.
