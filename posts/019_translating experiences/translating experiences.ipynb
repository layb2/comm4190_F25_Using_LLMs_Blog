A few of my classmates were tasked with learning how to juggle -- each pulling from a different knowledge resource, like a book, videos, LLM, or face-to-face interaction.

The classmate whose resource was an LLM discussed the struggles of learning through its instructions. When she was experiencing specific challenges, the LLM struggled to provide advice that helped solve them.

The discussion highlighted the limitations of LLMs in translating experiences that necessarily involve space and matter. They lack embodied knowledge, and their ability to advise is limited to the experiences that humans have written about and that ended up in the training dataset.

This offers plenty of interesting questions about what embodied experiences an LLM has or does not have "understanding" -- really, knowledge, of. For example, I would imagine there are plenty of resources online about how to learn to ride a bike. There are likely data in the LLM's training set that addresses the variety of niche pain points associated with the task.

Could the same be said for learning how to walk? Write? Make a PB&J?
