When I was in middle school, my teacher assigned each of us to write down instructions on how to complete tasks like making a peanut butter and jelly sandwich. 

I could not imagine that, today, the exercise would be so relevant in thinking about LLMs, or even that something that we refer to as "AI" would exist.

When the teacher went through each of our instructions, tiny lapses of knowledge produced major errors in his completion of the task. If someone's instructions said to "Get out two pieces of bread," he'd tear open a hole in the bag, instead of undoing the enclosure and opening it up. If someone's instructions said simply to "Put the peanut butter on the bread," he'd pick up the jar of peanut butter and place it on top of the loaf of bread.

I don't remember the exact lesson of the exercise -- precision of wording, I imagine -- but it is something that came to the forefront of my mind as we discussed embodied knowledge in class.

If an LLM became "embodied" -- that is, if a robot were given the ability to move about in the physical world, led by knowledge in the LLM's dataset -- I wonder if it would even be able to translate the knowledge to the physical world at all. If it could, would similar errors occur? How would training on videos and images change its ability to put its knowledge into an embodied context?
