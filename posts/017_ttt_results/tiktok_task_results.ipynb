## In light of the modern debate of remote versus in-office work, to what extent do the TikTok users in the comments on this video prefer the middle ground of hybrid work, compared to non-hybrid schedules (whether fully remote or fully in-office)?

#### The TikTok used in this study can be found at this link: https://www.tiktok.com/t/ZP8D5bEqu/. Below, you will find a series of graphs created as part of this exploration, assessments of each, and a reflection on what the compounded results reveal about hybrid vs. non-hybrid preferences among TikTok users who engaged with the video's comment section.


Figure 1: This plot shows the count of comments associated with each concept.



Figure 2: This plot displays the likes associated with each of the concepts, as well as the comments that do not fit into either of them.

#### The first plot displays the number of comments associated with Concepts 1 and 2. It is split almost half-and-half between them, demonstrating balance in the data. The second figure, which shows the distribution of likes across categories, reveals that Concept 2 (comments displaying a preference for a fully remote or fully in-office work schedule) was more popular. In terms of the research question, this preliminarily suggests that viewers of the comment section did not prefer comments advocating for the middle-ground hybrid option.



Figure 3: This plot shows the distribution of replies across the concepts, including a category for comments that do not fit into either one.

#### The third figure displays the split of replies across categories, revealing that Concept 2 had nearly triple the number of replies as Concept 1. The neither category was relatively equivalent to Concept 2, substantially larger than Concept 1. This means that even though there was a relatively equal balance between the number of comments under Concept 1 and the number of comments under Concept 2, the engagement tended to favor comments falling under Concept 2 (fully WFH/remote or fully in-office) than under Concept 1 (hybrid).



Figure 4: This plot displays how spread out the likes are and how concentrated they are in each of the categories.

#### The final figure offers the most revealing insights. Visualizing the spread of engagement and where concentrations lie provides information about virality in each of these categories. For example, the wider shape and shorter tail of the neither category shows that engagement with its comments was spread much more evenly than in the Concept categories, which are much more vertically stretched. The narrower shape and long tail of the Concept categories reveal that engagement within both concepts was spread across a narrower range than for the neither category. This effect is most pronounced for Concept 3; in other words, a small number of comments in this category received very high engagement in the form of likes.

#### In combination, these findings reveal that TikTok commenters did not show a strong preference for hybrid work. Though the number of comments supporting hybrid work versus non-hybrid (whether it was fully remote or fully in-office) was relatively balanced, the analysis reveals that user engagement favored comments that fell under Concept 2. In other words, the comments that were advocating for fully remote or fully in-office schedules received more likes and replies than the comments that advocated for hybrid work schedules or fell into the neither category. In conclusion, TikTok users who engaged with this posters video did not prefer the middle ground of hybrid work, but rather leaned toward non-hybrid schedules.
