The Semantle word of the day was "technology," let's see how an LLM does at guessing it!

#### USER
>Lets play a game of Semantle. Guess a word and I will send you its closeness, then guess again based on said closeness, and so on. Some preliminary info about the word of interest: The nearest word has a similarity of 83.32, the tenth-nearest has a similarity of 52.68 and the thousandth nearest word has a similarity of 29.54

#### ASSISTANT
>Great! Let's start guessing. My first guess is "universe." How close am I?

#### USER
>Similarity: 14.19. Proximity: cold

#### ASSISTANT
>Thanks for the information. Let's try another guess: "technology." How does that one compare?

Woah! Is this a fluke? After repeating the test multiple times, I received similar results; "technology" was always guessed in around only a few tries.

When I played the game, it was my first time (score of 38). I did not really know what to make of the information Semantle provided about the word of interest, so because I did not make (conscious) use of it, I decided to continue prompting and remove that part from it. Something interesting happened!

The number of guesses skyrocketed from a few to the tens, and there was no evidence of getting closer. The output became more rudimentary, repeating words or providing phrases (as opposed to single words) even after instruction not to. Aside from erring from the rules of the game, the guesses themselves, for the most part, resembled mine much more than the all-information prompt output did. 

Based on this, it can be posited that the informational "hint" which Semantle provides at the beginning means something to an LLM, which makes sense -- how quantitatively "close" a word is in relation to all other words is extremely useful information for a system built around such a concept. But, importantly, when that information was removed, and even when I continued to supply the similarity & proximity info associated with each of the LLM's guesses, it did not seem to improve. I initially thought it was just, by default, likely to guess "technology" regardless of the prompt/"hint," but this suggests otherwise. So why is that? Does Semantle reuse words, and the LLM has been trained to pair "technology" with that exact information about its semantic similarity to other words? Is there something more complex at hand?
