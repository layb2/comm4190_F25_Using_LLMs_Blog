Some interesting results that taught me more about how LLMs work were yielded from a simple “Yes, and” game with some classmates that I repeated with an LLM. The prompt was to start with a product (in this case, an oven) and build on it in each response, always starting with “Yes, and…”

The kinds of additions made by us were notably different than those made by the LLM. Our responses were imaginative, often unrealistic, and sometimes hesitant. Examples of them included:

**Yes, and it should be able to change in size at will.**

**Yes, and it should be able to teleport wherever we want it to go.**

**Yes, and you should be able to customize its color and lighting.**

**Oh, yes, and you should be able to... I don't know. I don't know, [make] like some designs for like, I don't know, new ovens or new kitchenware, I guess.**

**Yes, and it should be beautiful.**

I repeated this exercise by feeding our conversation (one response at a time) into an LLM. In contrast to our human-only game, each of the replies output by the LLM built neatly onto the previous prompt. For example, after we prompted
Yes, and you should be able to customize its color and lighting. The LLM output:

**Yes, and we can include an intuitive interface that allows users to easily change the oven's color and lighting effects, creating a personalized cooking ambiance.**

and when we said

**Yes, and you should be able to order food to teleport into your oven.**

the LLM output:

**Yes, and we could partner with local chefs and restaurants to provide a menu of gourmet dishes that can be instantly teleported into the oven, ensuring fresh and delicious meals at the press of a button.**

This exercise highlighted to me how the structured and goal-directed format of the training that LLMs receive inclines them toward elaboration rather than imagination; they are engineered to build and stay on track, whereas our human-only game jumped from concept to concept and, in the spirit of hesitation or pressure, sometimes included half-finished thoughts.


Image of human-only result (left) vs. LLM-involved result (right) generated by Copilot.

Our game led to a much more imaginative product, but the one involving the LLM led to one that is more easily realized. Can we say that one approach is fundamentally more “creative” than another? I also wonder to what extent the system’s moderation (in an effort to maintain a neutral, on-topic tone for the user) contributes to the predictability and relative lack of spontaneity in its output. A future blog post will explore how raising the temperature setting, which controls the randomness, affects the LLM’s responses in this game, and if a certain setting leads to output that more closely resembles our human game!
